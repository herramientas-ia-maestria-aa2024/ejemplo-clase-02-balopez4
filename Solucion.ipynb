```python
import spacy
```

```python
# usar un modelo propio de spaCy en español
nlp = spacy.load("es_core_news_sm")
```

```python
texto = open('discurso.txt','r',encoding='utf8')
```

```python
file=texto.read()
```

```python
file
```

```python
documento = nlp(file)
```

```python
# proceso para determinar las oraciones
for l in documento.sents:
    print(l)
```

```python
# proceso para obtener las entidades de un texto
for l in documento.sents:
    print(l.text, l.label_, len(l.text))
    
```
